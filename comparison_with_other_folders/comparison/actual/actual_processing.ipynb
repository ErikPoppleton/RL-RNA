```python
from re import sub

for ds in ['rfam', 'smalltestcase']:
    data = []
    with open('../expertRNA/Result_'+ds+'_NFE_4_branch_4_min_ENTRNA_Agave.csv_9_features_with_FE.csv') as f:
        culled_lines = f.readlines()[1::4]
        for l in culled_lines:
            l = l.split(',')
            print(l[0], l[1])
            data.append(sub('.ct|.txt','',l[1])+','+l[2]+','+l[3])
    if ds == 'smalltestcase':
        ds = 'matthews'
    with open('actual_'+ds+'.csv', 'w+') as f:
        f.write('\n'.join(data))
```
