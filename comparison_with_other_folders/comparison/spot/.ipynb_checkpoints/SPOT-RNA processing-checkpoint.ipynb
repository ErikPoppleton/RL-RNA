```python
# Convert all datasets into fasta format
from os import listdir

#testing dataset, ct format
for ds in ['rfam', 'matthews']:
    out = []
    for f in listdir('test_data/'+ds):
        out.append('>'+f[0:-3]+'\n') # <- maybe try 
        with open('test_data/'+ds+'/'+f) as g:
            l = g.readlines()
            out.append(l[1])

    with open('test_data/'+ds+'.fasta', 'w+') as o:
        o.write(''.join(out))

#initial training dataset, stockholm (st) format
out = []
for f in listdir('training_data/bpRNA_dataset/TR0'):
    out.append('>'+f.strip('.st')+'\n')
    with open('training_data/bpRNA_dataset/TR0/'+f) as g:
        l = g.readlines()
        out.append(l[3])

with open('training_data/bpRNA_dataset.fasta', 'w+') as o:
    o.write(''.join(out))
                 
#transfer learning dataset, wow, fasta files
out = []
for f in listdir('training_data/PDB_dataset/TR1_sequences/'):
    with open('training_data/PDB_dataset/TR1_sequences/'+f) as g:
        out.append(g.read())
    
with open('training_data/PDB_dataset.fasta', 'w+') as o:
    o.write(''.join(out))
```

```python
# build list of sequences in training
train_seqs = []

with open('training_data/PDB_dataset.fasta') as f:
    train_seqs.extend([l.strip() for l in f.readlines()[1::2]])
    
with open('training_data/bpRNA_dataset.fasta') as f:
    train_seqs.extend([l.strip() for l in f.readlines()[1::2]])
    
#print new file with the sequences excluded
for ds in ['rfam', 'matthews']:
    out = []
    
    with open('test_data/'+ds+'.fasta') as f:
        for i, l in enumerate(f.readlines()):
            if i%2 == 0:
                n = l.strip()
            else:
                if l.strip() in train_seqs:
                    print(n.strip('>'))
                else:
                    out.extend([n+'\n', l])
    with open('test_data/'+ds+'_cleaned.fasta', 'w+') as f:
        f.write(''.join(out))
```

```python
#Let's cut pseudoknots and format nicely
import subprocess
from re import sub
for name in ['rfam', 'mathews']:
    with open('spot_'+name+'.csv', 'w+') as output:
        output.write('Name,Seq,Structure_by_SPOT-RNA\n')
        for f in listdir('output+'+name):
            if 'bpseq' in f:
                process = subprocess.run(['python', '/home/erik/software/k2n_standalone/knotted2nested.py', '-f', 'bpseq', '-F', 'vienna', '-m', 'OA', '-o', 'BPS', 'output/'+f], capture_output=True, text=True)
                stdout = process.stdout.split('\n')
                seq = stdout[1]
                db = stdout[2]
                name = sub('.bpseq', '', f)
                output.write(','.join([name, seq, db]))
                output.write('\n')
```
